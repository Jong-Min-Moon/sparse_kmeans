```python
import numpy as np
from os import listdir
from os.path import isfile, join, isdir
import pandas as pd
```

```python
def get_rho(filename):
    for chunk in filename.replace(".csv", "").split("_"):
        if chunk.startswith('rho'):
            rho = float(chunk.replace("rho", ""))/10
    return(rho)

def get_p(filename):
    for chunk in filename.replace(".csv", "").split("_"):
        if chunk.startswith('p'):
            p = int(chunk.replace("p", ""))
    return(p)
```

```python
dir_list = [f for f in listdir("./") if isdir(join("./", f))]
del_list = [int(dir.replace("del","")) for dir in dir_list]
path_list = ['./del' + str(delta) + '/result/' for delta in del_list]
filelist_list = [ [f for f in listdir(path_del) if isfile(join(path_del, f))] for path_del in path_list ]

rho_list = []
for fileitem in filelist_list[0]:
    rho_list.append( get_rho(fileitem) )    
rho_list = np.unique( np.array(rho_list))

p_list = []
for fileitem in filelist_list[0]:
    p_list.append(
        [int(chunk.replace("p", "")) for chunk in fileitem.replace(".csv", "").split("_") if chunk.startswith('p')]
    )    
p_list = np.unique( np.array(p_list).flatten() )
```

```python
experiment_name = "anisotropic_cov_oracle"
df_list = []
for del_idx, filelist in enumerate(filelist_list):
    df_list.append( pd.DataFrame(columns=rho_list, index = p_list) )
    for filename in filelist:
        print(filename)
        df_list[del_idx].loc[get_p(filename), get_rho(filename)] = np.loadtxt(path_list[del_idx]+filename, delimiter=",", dtype=float).mean()
    df_list[del_idx].to_csv("./"+ experiment_name + "_del"+ str(del_list[del_idx]) + ".csv" )
```
