```python
import numpy as np
from os import listdir
import fnmatch

from os.path import isfile, join, isdir
import pandas as pd
```

```python
def get_rho(filename):
    for chunk in filename.replace(".csv", "").split("_"):
        if chunk.startswith('rho'):
            rho = float(chunk.replace("rho", ""))/10
    return(rho)

def get_p(filename):
    for chunk in filename.replace(".csv", "").split("_"):
        if chunk.startswith('p'):
            p = int(chunk.replace("p", ""))
    return(p)

def get_Delta(filename):
    for chunk in filename.replace(".csv", "").split("_"):
        if chunk.startswith('Delta'):
            Delta = int(chunk.replace("Delta", ""))
    return(Delta)
```

```python
filelist=  [f for f in listdir("./result/") if fnmatch.fnmatch(f, '*.csv')]

Delta_list = []
p_list = []
rho_list = []
for fileitem in filelist:
    rho_list.append( get_rho(fileitem) )    
    Delta_list.append( get_Delta(fileitem))  
    p_list.append( get_p(fileitem))  
rho_list = np.unique( np.array(rho_list))
Delta_list = np.unique( np.array(Delta_list))
p_list = np.unique( np.array(p_list))

print(rho_list)
print(Delta_list)
print(p_list)
```

```python



file_dict  = dict(zip(rho_list, [dict(
    zip(Delta_list, [dict(
        zip(p_list, ["" for p in p_list])
    ) for Delta in Delta_list])
) for rho in rho_list]))


for file_now  in filelist:
    rho_now =  get_rho(file_now)
    Delta_now =  get_Delta(file_now)
    p_now =  get_p(file_now)
    file_dict[rho_now][Delta_now][p_now] = file_now
```

```python
experiment_name = "anisotropic_cov_oracle_new"
df_list = dict(
    zip(rho_list, [pd.DataFrame(columns=Delta_list, index =  p_list) for rho in rho_list])
    )
```

```python

for rho in rho_list:
    rho_dict = file_dict[rho]
    for Delta in Delta_list:
        rho_Delta_dict = rho_dict[Delta]
        for p in p_list:
            file_name_now = './result/' + 'rho' + str(int(rho*10)) + '_Delta' + str(Delta)+ '_p' +str(p)  +'.csv'
            try:
                mean_now = pd.read_csv(file_name_now, header = None).to_numpy().mean()
            except:
                mean_now = 0
            df_list[rho].loc[p, Delta] = mean_now
    df_list[rho].to_csv("./"+ experiment_name + "_rho"+ str(rho) + ".csv" )
```
